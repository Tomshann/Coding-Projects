As shown below the transformation applied to the vectors by the matrix causes a rotation and causes the image to also shrink in height and width.


```python

import numpy as np
import matplotlib.pyplot as plt

def Task1a(a:float, b:float):
    """
    This function transforms the vectors by the matrix to produce a brand new set of vectors
    ----
    parameters:
        a,b: These define the values of the terms a and b used in the matrix and vectors
    ----
    returns:
        newpart1,2,3,4 : the lists of newly transformed vectors
        a: the value of the a term

    """
    part1 = [np.array([[a],[a]]),np.array([[a],[-a]]),np.array([[-a],[-a]]),np.array([[-a],[a]])] #defining the vector points for the parts of the diagram
    part2 = [np.array([[1-a],[a-1]]),np.array([[-a/4],[a-1]]),np.array([[-a/4],[a/2]]),np.array([[1-a],[a/2]])]
    part3 = [np.array([[a/4],[a-1]]),np.array([[a-1],[a/2]]),np.array([[a-1],[a-1]]),np.array([[a/4],[a/2]])]
    part4 = [np.array([[a-1],[-a/4]]),np.array([[a-1],[-a/2]]),np.array([[1-a],[-a/2]]),np.array([[1-a],[-a/4]])]
    A = (np.array([[b*1/b,1*1/b],[-1*1/b,1*1/b]])) #defining the matrix A that the vectors will be transformed by

    newpart1 = np.zeros((4,2))#creates a container for the new vector points
    i=0
    for each in part1:#iterates over the list of vectors
        newindex = [(A[0][0]*each[0][0]+A[0][1]*each[1][0]),(A[1][0]*each[0][0]+A[1][1]*each[1][0])] #Perform the multiplication of the two matrices by multiplying the first row of A by the first row of the vector point and then finding their sum, then multiplying the second row of A by the second row of the vector point and adding the two values. storing this result in a 2x1 matrix.
        newpart1[i] = newindex #adds the new vectors into a list
        i+=1
    newpart2 = np.zeros((4,2))
    i=0
    for each in part2:#iterates over the list of vectors
        newindex = [(A[0][0]*each[0][0]+A[0][1]*each[1][0]),(A[1][0]*each[0][0]+A[1][1]*each[1][0])] #Perform the multiplication of the two matrices by multiplying the first row of A by the first row of the vector point and then finding their sum, then multiplying the second row of A by the second row of the vector point and adding the two values. storing this result in a 2x1 matrix.
        newpart2[i] = newindex #adds the new vectors into a list
        i+=1
    newpart3 = np.zeros((4,2))
    i=0
    for each in part3:#iterates over the list of vectors
        newindex = [(A[0][0]*each[0][0]+A[0][1]*each[1][0]),(A[1][0]*each[0][0]+A[1][1]*each[1][0])] #Perform the multiplication of the two matrices by multiplying the first row of A by the first row of the vector point and then finding their sum, then multiplying the second row of A by the second row of the vector point and adding the two values. storing this result in a 2x1 matrix.
        newpart3[i] = newindex #adds the new vectors into a list
        i+=1
    newpart4 = np.zeros((4,2))
    i=0
    for each in part4:#iterates over the list of vectors
        newindex = [(A[0][0]*each[0][0]+A[0][1]*each[1][0]),(A[1][0]*each[0][0]+A[1][1]*each[1][0])] #Perform the multiplication of the two matrices by multiplying the first row of A by the first row of the vector point and then finding their sum, then multiplying the second row of A by the second row of the vector point and adding the two values. storing this result in a 2x1 matrix.
        newpart4[i] = newindex #adds the new vectors into a list
        i+=1
    return [newpart1,newpart2,newpart3,newpart4,a] #returns the newly transformed vectors and the value of a
   


def task1b(newpart1:list,newpart2:list,newpart3:list,newpart4:list,a:int):
    """
    This function produces the drawings of the vector parts returned from part1a and creates a comparison figure of the original vectors
    ----
    parameters:
        newpart1,2,3,4: define the new vector points to be plotted on the figure
        a: defines the value of the a term, this is used for plotting the original vectors
    ----
    returns:
        2 figures display the before and after plots of the vectors undergoing transformation. 
    """
    plt.figure(1)#creating the first figure
    #defining all original x and y values to be plotted  
    part1 = [[a,a],[a,-a],[-a,-a],[-a,a]] 
    part2 = [[1-a,a-1],[-a/4,a-1],[-a/4,a/2],[1-a,a/2]]
    part3 = [[a/4,a-1],[a-1,a/2],[a-1,a-1],[a/4,a/2]]
    part4 = [[a-1,-a/4],[a-1,-a/2],[1-a,-a/2],[1-a,-a/4]]

    #defining the x and y coordinates of the points for the parts of the diagram    
    original1x = [part1[0][0],part1[1][0],part1[2][0],part1[3][0],part1[0][0]]
    original1y = [part1[0][1],part1[1][1],part1[2][1],part1[3][1],part1[0][1]]
    original2x = [part2[0][0],part2[1][0],part2[2][0],part2[3][0],part2[0][0]]
    original2y = [part2[0][1],part2[1][1],part2[2][1],part2[3][1],part2[0][1]]
    original3x = [part3[0][0],part3[1][0]]
    original3y = [part3[0][1],part3[1][1]]
    original32x = [part3[2][0],part3[3][0]]
    original32y = [part3[2][1],part3[3][1]]
    original4x = [part4[0][0],part4[1][0],part4[2][0],part4[3][0]]
    original4y = [part4[0][1],part4[1][1],part4[2][1],part4[3][1]]

    #plotting the vectors
    plt.plot(original1x,original1y,color="blue")
    plt.plot(original2x,original2y,color="blue")
    plt.plot(original3x,original3y,color="blue")
    plt.plot(original32x,original32y,color="blue")
    plt.plot(original4x,original4y,color="blue")
    plt.title("Before Transformation")#titles the figure
   

    
    
    plt.figure(2)#creates a new figure for the transformed vectors
    #defining all of the new x and y values to be plotted
    p1nxarray = [newpart1[0][0],newpart1[1][0],newpart1[2][0],newpart1[3][0],newpart1[0][0]]
    p1nyarray = [newpart1[0][1],newpart1[1][1],newpart1[2][1],newpart1[3][1],newpart1[0][1]]
    p2nxarray = [newpart2[0][0],newpart2[1][0],newpart2[2][0],newpart2[3][0],newpart2[0][0]]
    p2nyarray = [newpart2[0][1],newpart2[1][1],newpart2[2][1],newpart2[3][1],newpart2[0][1]]
    p3nxarray = [newpart3[0][0],newpart3[1][0]]
    p32nxarray = [newpart3[2][0],newpart3[3][0]]
    p3nyarray = [newpart3[0][1],newpart3[1][1]]
    p32nyarray = [newpart3[2][1],newpart3[3][1]]
    p4nxarray = [newpart4[0][0],newpart4[1][0],newpart4[2][0],newpart4[3][0]]
    p4nyarray = [newpart4[0][1],newpart4[1][1],newpart4[2][1],newpart4[3][1]]
    
    #plotting the vectors
    plt.plot(p4nxarray,p4nyarray,color="red")
    plt.plot(p3nxarray,p3nyarray,color="red")
    plt.plot(p32nxarray,p32nyarray,color="red")
    plt.plot(p2nxarray,p2nyarray,color="red")
    plt.plot(p1nxarray,p1nyarray,color="red")
    plt.title("After Transformation") #titles the figure
    plt.show()#displays the figures

x= Task1a(4.0,3.0) #takes the parameters returned from exercise 1a and stores them in an array
task1b(x[0],x[1],x[2],x[3],x[4]) #calls the exercise1b function using the parameters returned in part a
```

As seen by the plotted figures below the smaller the value of the step size h used, the more accurate Euler's method approximates the soultion, as smaller h sizes are selected it is most likely that the approximation will become near identical to the exact solution as presented above. Also as shown by the figure the Euler's method approximates below the actual solution.

The percentage errors for each approximation are as follows(all errors calculated at x=1):

h=1 error=2.15%

h=0.2 error=0.03%

h=0.05 error= 0.0085%

Hence, the above results show that the smaller the step size chosen the more accurate that the approximation becomes.



```python
import numpy as np
import matplotlib.pyplot as plt
import math
def exercise2a():
    """
        This function calculates the approximate y values for the differential equation over the interval [0,5] for step sizes 1,0.2,0.05
        ----
        returns:
            step1,02,005approximations: these are the approximate y values for each value of Xn
    """
    #initial conditions for the differential equations
    y=-3
    x=0
    step1approximations=[y] #defines an array to store the values of the approximations
    for i in range(0,5,1): #iterates over the interval [0,5]
        y = y+ 1*gradient(x,y)  #calculates the next y value 
        x+=1 #increments x by the step value
        step1approximations.append(y) #adds the y value to the list

    x=0
    y=-3
    step02approximations=[y]
    for i in range(25): #this function iterates over the equivalent of [0,5] as 25*0.2 = 5
        y = y+ 0.2*gradient(x,y)
        x+=0.2
        step02approximations.append(y)
    
    x=0
    y=-3
    step005approximations=[y]
    for i in range(100): #this function iterates over the equivalent of [0,5] as 100*0.05 = 5
        y = y+ 0.05*gradient(x,y)
        x+=0.05
        step005approximations.append(y)
    
    return [step1approximations,step02approximations,step005approximations] #Returns the approximate y value lists
     
def gradient(x:float,y:float): #defines the differential equation
    return (x + (y/5))



def exercise2b(s1,s02,s005):
    """
    This function plots the approximations against the exact solution to the differential equation.
    ----
    parameters:
        s1,s02,s005: these are the lists of approximate y values
    ----
    returns:
        a figure of the approximations plotted against the exact solution

    """
    x= np.array(range(6)) #creates an array of size 5
    y= (-5*x) +22*(math.e**(x/5))-25 #defines the exact solution to the differential equation
    #creating arrays for the x values
    s1x = np.array(range(6)) 
    s02x = np.array(range(26))/5 
    s005x = np.array(range(101))/20
    #plotting the lines
    plt.plot(x,y,label = 'exact solution')
    plt.plot(s1x,s1,label = 'h = 1')
    plt.plot(s02x,s02,label='h = 0.2')
    plt.plot(s005x,s005,label='h= 0.05')
    plt.legend()
    plt.show()


x = exercise2a()
exercise2b(x[0],x[1],x[2])
```
